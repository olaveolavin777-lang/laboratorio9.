Generame un codigo mas completo del análisis simple de las ventas de un pequeño negocio, incluyendo la
creación de un DataFrame, el uso de operaciones de NumPy y la visualización de los resultados
con Matplotlib. Graficas detalladas y complementado
Los estudiantes trabajarán con los siguientes datos de ventas mensuales:
Datos de ventas mensuales (en miles)
data = {“Mes": ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto",
"Septiembre", "Octubre", "Noviembre", "Diciembre"], "Ventas": [25, 30, 28, 35, 40, 50, 55, 60,
70, 80, 75, 90],}

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Datos de ventas mensuales proporcionados
data = {"Mes": ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto",
"Septiembre", "Octubre", "Noviembre", "Diciembre"], "Ventas": [25, 30, 28, 35, 40, 50, 55, 60,
70, 80, 75, 90]}

# Crear un DataFrame de Pandas
df_ventas = pd.DataFrame(data)

print("DataFrame de Ventas:")
display(df_ventas)
```

### **Análisis de Ventas con NumPy**
Ahora, realizaremos algunas operaciones estadísticas con NumPy para entender mejor las ventas.

```python
# Convertir la columna 'Ventas' a un array de NumPy para operaciones
ventas_np = np.array(df_ventas['Ventas'])

# Calcular estadísticas básicas usando NumPy
venta_total = np.sum(ventas_np)
venta_promedio = np.mean(ventas_np)
venta_maxima = np.max(ventas_np)
venta_minima = np.min(ventas_np)
desviacion_estandar = np.std(ventas_np)

print(f"Venta Total Anual: {venta_total} mil unidades")
print(f"Venta Promedio Mensual: {venta_promedio:.2f} mil unidades")
print(f"Venta Máxima en un Mes: {venta_maxima} mil unidades")
print(f"Venta Mínima en un Mes: {venta_minima} mil unidades")
print(f"Desviación Estándar de Ventas: {desviacion_estandar:.2f} mil unidades")

# Calcular el cambio porcentual de ventas mes a mes
# np.diff calcula la diferencia entre elementos consecutivos
cambio_ventas = np.diff(ventas_np)

# Para calcular el porcentaje, necesitamos dividir por la venta del mes anterior
porcentaje_cambio = (cambio_ventas / ventas_np[:-1]) * 100

# Añadir estos datos al DataFrame para visualización
df_ventas['Cambio Ventas'] = [np.nan] + list(cambio_ventas) # El primer mes no tiene cambio anterior
df_ventas['% Cambio Ventas'] = [np.nan] + list(porcentaje_cambio)

print("\nDataFrame con Cambios de Ventas:")
display(df_ventas)
```

### **Visualización de Resultados con Matplotlib y Seaborn**
Ahora, crearemos gráficos detallados para visualizar las tendencias y patrones de ventas.

```python
plt.figure(figsize=(14, 7))
sns.barplot(x='Mes', y='Ventas', data=df_ventas, palette='viridis')
plt.title('Ventas Mensuales a lo largo del Año', fontsize=16)
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Ventas (en miles)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(x='Mes', y='Ventas', data=df_ventas, marker='o', color='purple', linewidth=2)
plt.title('Tendencia de Ventas Mensuales', fontsize=16)
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Ventas (en miles)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.barplot(x='Mes', y='% Cambio Ventas', data=df_ventas, palette='coolwarm')
plt.title('Porcentaje de Cambio en Ventas Mes a Mes', fontsize=16)
plt.xlabel('Mes', fontsize=12)
plt.ylabel('% Cambio de Ventas', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.axhline(0, color='grey', linestyle='--') # Línea de referencia para el 0% de cambio
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 10))
# Generar colores dinámicamente
colores = plt.cm.tab20.colors[:len(df_ventas)]
plt.pie(df_ventas['Ventas'], labels=df_ventas['Mes'], autopct='%1.1f%%', startangle=90, colors=colores, pctdistance=0.85)
plt.title('Distribución Porcentual de Ventas por Mes', fontsize=16)
plt.axis('equal') # Esto asegura que el círculo sea un círculo.
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_ventas['Ventas'], bins=5, kde=True, color='skyblue')
plt.title('Distribución de Frecuencia de Ventas Mensuales', fontsize=16)
plt.xlabel('Ventas (en miles)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Realizar un análisis completo y detallado de las calificaciones de estudiantes en diferentes
asignaturas, incluyendo la creación de un DataFrame, el uso de operaciones de NumPy y la
visualización de los resultados con Matplotlib.
Los estudiantes trabajarán con los siguientes datos de calificaciones:
Datos de calificaciones de estudiantes
data = {"Estudiante": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah",
"Ian", "Jack"], "Matemáticas": [85, 92, 78, 90, 88, 76, 95, 89, 82, 94], "Ciencias": [90, 85, 80,
88, 92, 75, 95, 87, 81, 93], "Historia": [78, 82, 88, 90, 85, 80, 92, 84, 79, 91],}

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Datos de calificaciones de estudiantes proporcionados
data_calificaciones = {
    "Estudiante": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ian", "Jack"],
    "Matemáticas": [85, 92, 78, 90, 88, 76, 95, 89, 82, 94],
    "Ciencias": [90, 85, 80, 88, 92, 75, 95, 87, 81, 93],
    "Historia": [78, 82, 88, 90, 85, 80, 92, 84, 79, 91]
}

# Crear un DataFrame de Pandas
df_calificaciones = pd.DataFrame(data_calificaciones)

print("DataFrame de Calificaciones:")
display(df_calificaciones)
```

### **Análisis Estadístico con NumPy**
Ahora, realizaremos algunas operaciones estadísticas con NumPy para entender mejor las calificaciones por asignatura y por estudiante.

```python
# Seleccionar solo las columnas de asignaturas para el análisis numérico
asignaturas = ['Matemáticas', 'Ciencias', 'Historia']
calificaciones_np = df_calificaciones[asignaturas].values

# Calcular estadísticas generales por asignatura
print("\nEstadísticas por Asignatura:")
for i, asignatura in enumerate(asignaturas):
    calificaciones_asignatura = calificaciones_np[:, i]
    print(f"--- {asignatura} ---")
    print(f"  Promedio: {np.mean(calificaciones_asignatura):.2f}")
    print(f"  Mínima: {np.min(calificaciones_asignatura)}")
    print(f"  Máxima: {np.max(calificaciones_asignatura)}")
    print(f"  Desviación Estándar: {np.std(calificaciones_asignatura):.2f}")
    print(f"  Mediana: {np.median(calificaciones_asignatura)}")

# Calcular el promedio de calificaciones por estudiante
df_calificaciones['Promedio General'] = np.mean(calificaciones_np, axis=1)

# Identificar el estudiante con la calificación más alta y más baja en promedio
mejor_estudiante_idx = np.argmax(df_calificaciones['Promedio General'])
peor_estudiante_idx = np.argmin(df_calificaciones['Promedio General'])

print(f"\nEstudiante con el mejor promedio: {df_calificaciones.loc[mejor_estudiante_idx, 'Estudiante']} ({df_calificaciones.loc[mejor_estudiante_idx, 'Promedio General']:.2f})")
print(f"Estudiante con el peor promedio: {df_calificaciones.loc[peor_estudiante_idx, 'Estudiante']} ({df_calificaciones.loc[peor_estudiante_idx, 'Promedio General']:.2f})")

print("\nDataFrame de Calificaciones con Promedio General:")
display(df_calificaciones)
```

### **Visualización de Resultados con Matplotlib y Seaborn**
Ahora, crearemos gráficos detallados para visualizar las calificaciones y sus distribuciones.

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Estudiante', y='Promedio General', data=df_calificaciones.sort_values(by='Promedio General', ascending=False), palette='coolwarm')
plt.title('Promedio General de Calificaciones por Estudiante', fontsize=16)
plt.xlabel('Estudiante', fontsize=12)
plt.ylabel('Promedio de Calificaciones', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_calificaciones[asignaturas], palette='viridis')
plt.title('Distribución de Calificaciones por Asignatura', fontsize=16)
plt.xlabel('Asignatura', fontsize=12)
plt.ylabel('Calificación', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
df_melted = df_calificaciones.melt(id_vars=['Estudiante', 'Promedio General'], value_vars=asignaturas, var_name='Asignatura', value_name='Calificacion')

plt.figure(figsize=(14, 7))
sns.lineplot(x='Asignatura', y='Calificacion', hue='Estudiante', data=df_melted, marker='o')
plt.title('Calificaciones Individuales por Asignatura', fontsize=16)
plt.xlabel('Asignatura', fontsize=12)
plt.ylabel('Calificación', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
sns.heatmap(df_calificaciones[asignaturas].corr(), annot=True, cmap='cividis', fmt='.2f')
plt.title('Matriz de Correlación entre Asignaturas', fontsize=16)
plt.tight_layout()
plt.show()
```
